```python
import numpy as np
import os
import cv2
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../dataset/Raw_dataset/SN_goinka/2Q__ (10).jpg")
img.shape
```

```python
accepted_files = []
count = 0
for dirpath , dirname , files in os.walk(r"../dataset/Raw_dataset/"):
    for file in files:
        try:
            filepath = os.path.join(dirpath, file)
            img = cv2.imread(filepath)
            
            if img.shape[0] > 2000 or img.shape[1] > 2000:
                print(file)
                print(dirpath.split("/")[-1])
                print(count)
                accepted_files.append(filepath)
                count = count + 1
                plt.figure()
                plt.imshow(img)

        except:
            pass
#         break
#     break
```

```python
gray = cv2.cvtColor(img , cv2.COLOR_RGB2GRAY )
```

```python
gray.shape
```

```python
plt.imshow(gray,cmap = 'gray')
```

```python
face_cascade = cv2.CascadeClassifier("./opencv/haarcascades/haarcascade_frontalface_default.xml")
```

```python
face = face_cascade.detectMultiScale(gray)
```

```python
x, y, w, h = face[0]
```

```python
face_img = cv2.rectangle(img , (x, y), (x + w , y+ h) , (255, 0, 0), 2)
```

```python
plt.imshow(face_img)
```

```python
face_crop_gray = gray[y:y+h , x:x+w]
```

```python
plt.imshow(face_crop_gray)
```

```python
face_crop_gray.shape
```

```python
eye_cascade = cv2.CascadeClassifier("./opencv/haarcascades/haarcascade_eye_tree_eyeglasses.xml")
```

```python
eye = eye_cascade.detectMultiScale(face_crop_gray)
```
